```python
# Importación de bibliotecas necesarias
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA

# 1. Cargar los datos
data = {
    'City': ['Bogotá', 'Medellín', 'Cali', 'Barranquilla', 'Cartagena', 'Pereira', 'Santa Marta', 'Manizales', 'Villavicencio', 'Pasto',
             'Montería', 'Valledupar', 'Neiva', 'Popayán', 'Armenia', 'Sincelejo', 'Tunja', 'Florencia', 'Riohacha', 'Quibdó',
             'San Andrés', 'Yopal', 'Leticia', 'Mocoa', 'Mitú'],
    'GDP (USD Billion)': [103.5, 44.1, 22.4, 16.8, 10.5, 6.2, 4.0, 3.8, 3.5, 3.2, 3.0, 2.8, 2.5, 2.3, 2.1, 2.0, 1.8, 1.7, 1.5, 1.3,
                          1.2, 1.1, 1.0, 0.8, 0.7],
    'Population (Millions)': [7.18, 2.57, 2.23, 1.23, 1.03, 0.48, 0.52, 0.43, 0.50, 0.45, 0.49, 0.47, 0.35, 0.33, 0.30, 0.28, 0.25, 0.20,
                              0.22, 0.13, 0.08, 0.15, 0.05, 0.04, 0.01],
    'Unemployment Rate (%)': [10.5, 11.2, 13.8, 12.4, 10.9, 12.0, 11.6, 10.7, 13.0, 12.9, 13.5, 14.8, 14.1, 15.2, 13.3, 16.5, 10.0, 17.5,
                              15.7, 18.2, 14.0, 11.5, 13.6, 15.0, 20.0],
    'Average Age': [32, 31, 30, 29, 30, 32, 29, 32, 30, 31, 29, 28, 30, 31, 32, 29, 31, 28, 27, 26, 27, 29, 26, 28, 25],
    'Women (%)': [52, 53, 52, 51, 51, 52, 51, 53, 51, 52, 51, 51, 52, 52, 53, 51, 52, 51, 51, 52, 50, 51, 51, 52, 51],
    'Men (%)': [48, 47, 48, 49, 49, 48, 49, 47, 49, 48, 49, 49, 48, 48, 47, 49, 48, 49, 49, 48, 50, 49, 49, 48, 49],
    'Budget (USD Billion)': [18.0, 7.5, 4.2, 3.1, 2.8, 1.3, 0.9, 0.8, 0.8, 0.7, 0.7, 0.6, 0.6, 0.5, 0.5, 0.5, 0.4, 0.4, 0.3, 0.3, 0.2,
                             0.2, 0.1, 0.1, 0.05]
}


df = pd.DataFrame(data)


media = df[['GDP (USD Billion)', 'Population (Millions)', 'Unemployment Rate (%)', 'Average Age', 'Women (%)', 'Men (%)', 'Budget (USD Billion)']].mean()
mediana = df[['GDP (USD Billion)', 'Population (Millions)', 'Unemployment Rate (%)', 'Average Age', 'Women (%)', 'Men (%)', 'Budget (USD Billion)']].median()
desviacion_estandar = df[['GDP (USD Billion)', 'Population (Millions)', 'Unemployment Rate (%)', 'Average Age', 'Women (%)', 'Men (%)', 'Budget (USD Billion)']].std()

# Cálculo de la moda para datos categóricos (en este caso, las ciudades)
moda_city = df['City'].mode()[0]
moda_city_count = df['City'].value_counts()[moda_city]

# Mostrar las estadísticas calculadas
print("Media de cada variable:")
print(media)
print("\nMediana de cada variable:")
print(mediana)
print("\nDesviación estándar de cada variable:")
print(desviacion_estandar)
print(f"\nLa moda de la columna 'City' es: {moda_city} con {moda_city_count} repeticiones.")



X1 = df['GDP (USD Billion)']
X2 = df['Population (Millions)']


X1_media = X1.mean()
X2_media = X2.mean()

# Calcular la covarianza utilizando la fórmula manualmente
covarianza_manual = sum((X1 - X1_media) * (X2 - X2_media)) / len(X1)

print(f"La covarianza calculada manualmente es: {covarianza_manual}")


X1 = df['GDP (USD Billion)']
X2 = df['Population (Millions)']


X1_media = X1.mean()
X2_media = X2.mean()


numerador = sum((X1 - X1_media) * (X2 - X2_media))


denominador = np.sqrt(sum((X1 - X1_media) ** 2) * sum((X2 - X2_media) ** 2))


correlacion_manual = numerador / denominador

print(f"La correlación calculada manualmente es: {correlacion_manual}")

#punto 2


# Usar solo las columnas GDP y Population para el PCA
X = df[['GDP (USD Billion)', 'Population (Millions)']]

# Estándar de los datos
X_meaned = X - np.mean(X, axis=0)

# 2.1: Matriz de covarianza
cov_matrix = np.cov(X_meaned, rowvar=False)
print("Matriz de covarianza:\n", cov_matrix)

# 2.2: Cálculo de los eigenvalores y eigenvectores
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
print("\nEigenvalores:\n", eigenvalues)
print("\nEigenvectores:\n", eigenvectors)

# 2.3: Varianza explicada por cada eigenvalue
explained_variance = eigenvalues / np.sum(eigenvalues)
print("\nVarianza explicada:\n", explained_variance)

# 2.4: Eigenvector correspondiente al mayor eigenvalue
principal_eigenvector = eigenvectors[:, np.argmax(eigenvalues)]
print("\nPrincipal eigenvector:\n", principal_eigenvector)

# 2.5: Matriz proyectada
X_pca = X_meaned.dot(principal_eigenvector)
print("\nMatriz proyectada:\n", X_pca)

# 2.6: Error o diferencia entre la matriz original y la proyectada
reconstructed_X = np.outer(X_pca, principal_eigenvector) + np.mean(X, axis=0)
error = X - reconstructed_X
print("\nError:\n", error)

# 2.7: Pintar las ciudades en 1 dimensión
plt.figure(figsize=(8,6))
plt.scatter(X_pca, np.zeros_like(X_pca), color='red')
plt.title('Proyección PCA en 1 dimensión')
plt.xlabel('Proyección sobre el primer componente principal')
plt.show()

# 2.8: Pintar todas las ciudades en 2 dimensiones
pca_2d = PCA(n_components=2)
X_pca_2d = pca_2d.fit_transform(X)
plt.figure(figsize=(8,6))
plt.scatter(X_pca_2d[:, 0], X_pca_2d[:, 1], color='blue')
for i, city in enumerate(df['City']):
    plt.text(X_pca_2d[i, 0], X_pca_2d[i, 1], city)
plt.title('Proyección PCA en 2 dimensiones')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.grid()
plt.show()

```
